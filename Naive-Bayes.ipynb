```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import scipy 
import seaborn as sns 
```

```python
df = pd.read_csv("./datasets/Prostate_Cancer.csv")
df.head()
```

```python
df.diagnosis_result
```

```python
df.isnull().values.any()
```

```python
df["diagnosis_result"].value_counts()
```

```python
df = df.drop(labels=['id'], axis=1)
df.head()
```

```python
from sklearn.model_selection import train_test_split
x = df.drop(labels = "diagnosis_result" , axis = 1)

y = df["diagnosis_result"]
y
x.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.3, random_state=80)
print(np.around(len(x_train)/len(df)*100,2), '% data is in training set')
print(np.around(len(x_test)/len(df)*100,2), '% data is in testing set')
x_train.head()
```

```python
from sklearn.impute import SimpleImputer

replacement = SimpleImputer(missing_values=0, strategy='median')

cols = x_train.columns
x_train = pd.DataFrame(replacement.fit_transform(x_train))
x_test = pd.DataFrame(replacement.fit_transform(x_test))
```

```python
x_train.columns = cols
x_test.columns = cols
```

```python
x_train.head()
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

```python
model.fit(x_train , y_train)
```

```python
y_pred_train = model.predict(x_train)
```

```python
from sklearn import metrics
```

```python
y_pred_test = model.predict(x_test)
```

```python
print("Model Accuracy:", np.around(metrics.accuracy_score(y_train, y_pred_train)*100, 2), '%')
```

```python
print("Model Accuracy:",np.around(metrics.accuracy_score(y_test, y_pred_test)*100, 2),"%")
```

```python
confusion = metrics.confusion_matrix(y_test, y_pred_test, labels=['M', 'B'])
confusion_df = pd.DataFrame(confusion, index = ["Real M" , "Real B"], columns = ["Predict M" , "Predict  B"])
sns.heatmap(confusion_df , annot=True)
```

```python
from sklearn.metrics import f1_score, recall_score, precision_score
f1_score(y_test, y_pred_test, average='macro')
```

```python
f1_score(y_test, y_pred_test, average='micro')
```

```python
precision_score(y_test, y_pred_test, average='weighted')
```
